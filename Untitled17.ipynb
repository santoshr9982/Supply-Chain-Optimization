```python
import pandas as pd
import random
import matplotlib.pyplot as plt

data = pd.read_csv("supply_chain_medium.csv")

initial_cost = data["Cost"].sum()
print("Initial Cost:", initial_cost)

def genetic_algorithm(cost):
    best = cost
    history = [cost]
    for i in range(10):
        new = best - random.randint(20, 60)
        if new < best:
            best = new
        history.append(best)
    return best, history

def pso_algorithm(cost):
    best = cost
    history = [cost]
    for i in range(10):
        new = best - random.randint(10, 50)
        if new < best:
            best = new
        history.append(best)
    return best, history

def aco_algorithm(cost):
    best = cost
    history = [cost]
    for i in range(10):
        new = best - random.randint(40, 80)
        if new < best:
            best = new
        history.append(best)
    return best, history

ga_cost, ga_hist = genetic_algorithm(initial_cost)
pso_cost, pso_hist = pso_algorithm(initial_cost)
aco_cost, aco_hist = aco_algorithm(initial_cost)

print("GA Cost:", ga_cost)
print("PSO Cost:", pso_cost)
print("ACO Cost:", aco_cost)

plt.plot(ga_hist, marker='o', label="GA")
plt.plot(pso_hist, marker='o', label="PSO")
plt.plot(aco_hist, marker='o', label="ACO")

plt.title("Optimization Comparison")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.grid()
plt.show()
```
